```python
import torch
import cv2
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

# Load YOLOv5 model
model = torch.hub.load('ultralytics/yolov5', 'yolov5n', pretrained=True)

# Open a connection to the camera (0 for the default camera)
cap = cv2.VideoCapture(0)

while True:
    # Capture frame-by-frame
    try:
        ret, frame = cap.read()
        # if not ret:
        #     print("Failed to grab frame")
        #     break

        # Convert the frame (BGR to RGB)
        img_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        # Perform inference
        results = model(img_rgb)

        # Get predictions
        predictions = results.xyxy[0]  # Bounding boxes (x1, y1, x2, y2, confidence, class)

        # Draw rectangles for each prediction
        for pred in predictions:
            # Get coordinates and confidence
            x1, y1, x2, y2, conf, cls = pred.tolist()
            
            # Filter for only 'person' class (class 0)
            if cls == 0:  # Only keep predictions for 'person'
                # Create a rectangle on the original frame
                cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), np.random.rand(3,), 2)
                # Optionally, add a label with the confidence score
                cv2.putText(frame, f'{conf:.2f}', (int(x1), int(y1) - 10), 
                            cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2)

        # Show the resulting frame with detections
        cv2.imshow('YOLOv5 Human Detection', frame)


        if cv2.waitKey(1) & 0xFF == 0xd:
            break
    except Exception as e:
        print(e.args)
        break
# Release the camera and close windows
cap.release()
cv2.destroyAllWindows()
```
